#Preprocesamiento
#### - Formato








```python
#Transformaciónes del DataFrame ---> Arreglar el formato de fecha en el DataFrame y (era texto no datetime)
df_y['Date'] = pd.to_datetime(df_y['Date'], dayfirst=True, errors='raise')
df_y
```

```python
#Asegurar que DataFrame x - z tienen el formato de fecha tipo datetime
df_x['Date'] = pd.to_datetime(df_x['Date'])
df_z['Date'] = pd.to_datetime(df_z['Date'])
```

```python
df_x = df_x[['Date', 'Price']]
df_x.head()
```

```python
df_z = df_z[['Date', 'Price']]
df_z.head()
```

#Hallazgos iniciales

####Todos los df una poseen columna de Date y Price, no hay valores nulos dentro de las columnas.
####Date esta en formato datetime y Price es tipo float64.
####Las fechas en los df X, Y y Z son diferentes, por lo que se supone que los datos provienen de diferentes períodos de tiempo.

#Visualización de los datos

```python
plt.figure(figsize=(12, 6))
plt.plot(pd.to_datetime(df_x['Date']), df_x['Price'], label='Price')
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('X Price over Time')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(pd.to_datetime(df_y['Date']), df_y['Price'], label='Price')
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Y Price over Time')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(pd.to_datetime(df_z['Date']), df_z['Price'], label='Price')
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Z Price over Time')
plt.legend()
plt.tight_layout()
plt.show()
```
